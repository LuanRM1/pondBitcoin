# Exploração de Dados do Preço de Avalanche (AVAX)

### 1. Coleta de Dados Históricos (5 anos)

```python

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

avax = yf.Ticker("AVAX-USD")
avax_data = avax.history(period="5y")

avax_data.head()
```

### 2. Resumo Estatístico

```python

# Resumo estatístico dos dados
avax_data.describe()
```

### 3. Visualização dos Preços de Fechamento

```python

plt.figure(figsize=(14, 7))
plt.plot(avax_data['Close'], label='Preço de Fechamento')
plt.title('Histórico de Preços de Fechamento do Avalanche (AVAX)')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento (USD)')
plt.legend()
plt.show()
```

### 4. Análise de Correlações

```python

correlation_matrix = avax_data[['Open', 'High', 'Low', 'Close']].corr()
correlation_matrix
```

### 5. Análise de Distribuição

```python

plt.figure(figsize=(10, 5))
plt.hist(avax_data['Close'], bins=50, color='blue', edgecolor='black')
plt.title('Distribuição dos Preços de Fechamento do AVAX')
plt.xlabel('Preço de Fechamento (USD)')
plt.ylabel('Frequência')
plt.show()
```
